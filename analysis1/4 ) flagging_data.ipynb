```python
import pandas as pd

data = pd.read_csv('data/with_track_language.csv')
data.head(5)
```

```python
import time
now = time.time()
for index,row in data.iterrows():
    if data.at[index,"Track_Language"] in data.at[index,"Region_Language"]:
        data.at[index,"flag"] = 1
    else:
        data.at[index,"flag"] = 0
end = time.time()
print(end-now,"time diff")
```
